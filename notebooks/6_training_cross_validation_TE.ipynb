## Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,root_mean_squared_error,r2_score

target_encoding = ['Manufacturer' ,'Model' ,'Category' , 'Cylinders' ,'Fuel type' ,'Color' ,'Airbags']
numerical_columns = ['Levy','Engine volume', 'Mileage' , 'Age' ,'Manufacturer' ,'Model' ,'Category' , 'Cylinders' ,'Fuel type' ,'Color' ,'Airbags']
```

```python
accuracy_linear_regression = {'Linear Regression RMSE' : [],
                            'Linear Regression MSE' : [],
                            'Linear Regression R2 score' : []}

accuracy_Random_forest = {'Random Forest RMSE' : [],
                            'Random Forest MSE' : [],
                            'Random Forest R2 score' : []}
```

```python
skfold = StratifiedKFold(n_splits=10 , shuffle=True , random_state=42)
```

```python
for train_idx , val_idx in skfold.split(x_train,y_train):
    x_train_fold , y_train_fold = x_train.iloc[train_idx] , y_train.iloc[train_idx]
    x_validation_fold , y_validation_fold = x_train.iloc[val_idx] , y_train.iloc[val_idx]

    x_train_fold_target = x_train_fold.copy()
    x_train_fold_target['price'] = y_train_fold

    global_mean = x_train_fold_target['price'].mean()

    for col in target_encoding:
        mean_encoded = x_train_fold_target.groupby(col)['price'].mean()

        x_train_fold[col] = x_train_fold[col].map(mean_encoded).fillna(global_mean)
        x_validation_fold[col] = x_validation_fold[col].map(mean_encoded).fillna(global_mean)
    
    scaler_cross = StandardScaler()

    x_train_fold[numerical_columns] = scaler_cross.fit_transform(x_train_fold[numerical_columns])
    x_validation_fold[numerical_columns] = scaler_cross.transform(x_validation_fold[numerical_columns])

    x_train_fold = x_train_fold.drop(columns=['Gear box type' ,'Drive wheels' ,'Wheel'])
    x_validation_fold= x_validation_fold.drop(columns=['Gear box type' ,'Drive wheels' ,'Wheel'])


    lr = LinearRegression()
    lr.fit(x_train_fold,y_train_fold)
    y_pred = lr.predict(x_validation_fold)

    mse = mean_squared_error(y_validation_fold,y_pred)
    rmse = root_mean_squared_error(y_validation_fold,y_pred)
    r2 = r2_score(y_validation_fold,y_pred)

    accuracy_linear_regression['Linear Regression MSE'].append(mse)
    accuracy_linear_regression['Linear Regression RMSE'].append(rmse)
    accuracy_linear_regression['Linear Regression R2 score'].append(r2)


    rf =RandomForestRegressor()
    rf.fit(x_train_fold,y_train_fold)
    y_pred = rf.predict(x_validation_fold)

    mse = mean_squared_error(y_validation_fold,y_pred)
    rmse = root_mean_squared_error(y_validation_fold,y_pred)
    r2 = r2_score(y_validation_fold,y_pred)

    accuracy_Random_forest['Random Forest MSE'].append(mse)
    accuracy_Random_forest['Random Forest RMSE'].append(rmse)
    accuracy_Random_forest['Random Forest R2 score'].append(r2)
```

## Linear Regression Validation

```python
mean_mse = np.mean(accuracy_linear_regression['Linear Regression MSE'])
mean_Rmse = np.mean(accuracy_linear_regression['Linear Regression RMSE'])
mean_R2_score = np.mean(accuracy_linear_regression['Linear Regression R2 score'])

print(f'MSE score : {mean_mse}')
print(f'RMSE Score : {mean_Rmse}')
print(f'R2 Score Score : {mean_R2_score}')
```

## Random Forest Regresor

```python
mean_mse = np.mean(accuracy_Random_forest['Random Forest MSE'])
mean_Rmse = np.mean(accuracy_Random_forest['Random Forest RMSE'])
mean_R2_score = np.mean(accuracy_Random_forest['Random Forest R2 score'])

print(f'MSE Score : {mean_mse}')
print(f'RMSE Score : {mean_Rmse}')
print(f'R2 Score : {mean_R2_score}')
```
